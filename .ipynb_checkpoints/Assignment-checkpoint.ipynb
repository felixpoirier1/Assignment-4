## <span style="color:blue"> **Modules & Packages**</span>

```python
# Pandas: To manipulate DataFrames
import pandas as pd
# Seaborn & Matplotlib: Data Visualization
import seaborn as sns
import matplotlib.pyplot as plt
# Statsmodels: To estimate our regression model and perform heteroskedatsicity tests
import statsmodels.api as sm
import statsmodels.formula.api as smf
import statsmodels.tools.tools as smt
import statsmodels.stats.diagnostic as smd
# Numpy: For mathematical computations
import numpy as np
# YFinance to download financial data
import yfinance as yf
```

## <span style="color:blue"> **The Dataset**</span>

```python
# Creation of the DataFrame from multiple tables stored in an Excel file
sheet = ["Y variable", "x1-market", "x2-value", "x3-small"]
sets = []

data_dict = {}
for title in sheet:
    data_dict[title] = pd.DataFrame(pd.read_excel("TP2-Dataset.xlsx", skiprows = range(6), sheet_name=title)\
                .iloc[:][['Date', 'PX_LAST']])
    data_dict[title].set_index('Date', inplace=True)
    data_dict[title].index = pd.to_datetime(data_dict[title].index)
    data_dict[title][title] = np.log(data_dict[title]['PX_LAST']/data_dict[title]['PX_LAST'].shift(1))
    del data_dict[title]['PX_LAST']
    data_dict[title] = data_dict[title].dropna()
    sets += [data_dict[title]]

df = pd.concat(sets, axis = 1)
```

```python
name = ['Tech', 'Market', 'Value', 'Size']

ren = {sheet[i]: name for i, name in enumerate(name)}
df = df.rename(columns=ren)
```

```python
df.head()
```

```python
end = df.index[0]
start = df.index[-1]
```

```python
# Downloading the data from Yahoo Finance
oil = yf.download(tickers = "CL%3DF", start = start, end = end, interval = "1wk", ).iloc[:]['Close']

df['Oil'] = list(oil)[::-1]
df.Oil = np.log(df.Oil/df.Oil.shift(1))

df = df.dropna()
```

```python
df.cumsum().plot(figsize=(15, 10))
```

```python
df.corr()
```

## <span style="color:blue"> **Multivariate OLS Regression**</span>

```python
# Define dependant (y) and independent (x) variables
y = df[name[0]]
x = df[name[1:]]

# Add a constant to the predictor variable (intercept)
x = sm.add_constant(x)

# Fit the OLS model to the sample data
model_ols = sm.OLS(y, x).fit()

# View model summary (intercept, coefficient, r-squared)
print(model_ols.summary())
```

```python
# Find covariance between residuals and oil prices
df["Oil"].cov(model_ols.resid)
```

> The covariance between the our chosen instrument and the residuals of our OLS regression is relatively small. While this does not make the instrument valuable, as a covariance of 0 is what would be desired, it is a good indicator that the covariance between the instrument and the population error term may be small.

```python
df["Oil"].corr(model_ols.resid)
```
